# Association Rules: Cosmetic Recommendation

## Exploration

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = pd.read_csv('Cosmetics.csv')
dataset.head()
```

## Pre-processing

```python
dataset = dataset.drop(['Trans. '], axis=1)
dataset = dataset.astype(bool)
dataset.head()
```

## Model construction using association rules

### Generate all the frequent itemsets with support = 20%

```python
frequent_itemsets=apriori(dataset,min_support=0.2,use_colnames=True)
frequent_itemsets
```

### total number of frequent itemsets

```python
len(frequent_itemsets)
```

### Generate all the association rules with minimum support = 20% and minimum confidence = 70%

```python
my_rules = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
my_rules
```

## Model deployment

### Show rule with highest confidence

```python
my_rules.sort_values('confidence',ascending=False)
```

### Explain the meaning of this rule

Transaction ที่มี Mascara มีโอกาส 89.9 % ที่จะมี Eye Shadow อยู่ด้วย

Transaction ที่มี Eye Shadow มีโอกาส 84.25 % ที่จะมี Mascara อยู่ด้วย

Transaction ที่มี Lip Gloss มีโอกาส 72.65 % ที่จะมี Foundation อยู่ด้วย

### Explain how support, confidence and lift are calculated?

support = freq(A,B) / N

confidence = support / (antecedent support)

lift = confident / (consequent support)
