```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
## print shape of dataset with nb-rows and nb-columns
```

```python
## print the top5 records
df = dataset.copy()
df[0:5]
```

## Missing Values

```python
## Use heatmap tp show missing values features
df = dataset.copy()
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
## 2- step print the feature name and the percentage of missing values
df = dataset.copy()
df = df.isnull().mean().round(4)
dic = df[df > 0].to_dict()
for i in dic:
  print(f"{i} {dic[i]}  % missing values")
```

#### Temporal Variables(Eg: Datetime Variables)

```python
# list of variables that contain year information => Temporal variables
df = dataset.copy()
col = []
for i in df.head():
  if df[i].dtype == np.dtype('int64') or 'GarageYrBlt':
    try:
      df[i] = pd.to_datetime(df[i], format='%Y')
      col.append(i)
    except:
      pass
print(col)
```

```python
# let's explore the content of these year variables
df = dataset.copy()
for i in df.head():
  if df[i].dtype == np.dtype('int64') or 'GarageYrBlt':
    try:
      pd.to_datetime(df[i], format='%Y')
      print(i,df[i].unique())
      col.append(i)
    except:
      pass
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price
df = dataset.copy()
df.groupby("YrSold")['SalePrice'].median().plot(title="House Price vs YearSold", xlabel="Year Sold",ylabel="Median House Price")
```

```python
## Numerical variables are usually of 2 type, Continous variable and Discrete Variables (finite set of values < 25)
df = dataset.copy()
c = 0
l = []
for i in df.select_dtypes(include='number').head():
  if len(df[i].unique()) < 25:
    try:
      pd.to_datetime(df[i], format='%Y')
      continue
    except:
      pass
    c+=1
    l.append(i)
print(f"Discrete Variables Count: {c}")
```

```python
discrete_feature = l
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Lets Find the realtionship between Discrete Variables and Sale PRice
df = dataset.copy()
for i in discrete_feature:
  df.groupby(i)['SalePrice'].median().plot.bar(title=i,xlabel=i,ylabel="SalePrice")
  plt.show()
```

#### Continuous Variable

```python
df = dataset.copy()
c = 0
l = []
for i in df.select_dtypes(include='number').head():
  if len(df[i].unique()) > 25 and i != "Id":
    try:
      pd.to_datetime(df[i], format='%Y')
      continue
    except:
      pass
    c+=1
    l.append(i)
print(f"Continuous feature Count {c}")
continuous_feature = l.copy()
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution
df = dataset.copy()
for i in continuous_feature:
  df[i].plot.hist(title=i,x=df[i].unique(),y=df[i].value_counts(),xlabel=i,ylabel="Count",grid=True)
  plt.show()
```

### Categorical Variables
