```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
bank_data = pd.read_csv('bank-data.csv')
bank_data.head()
```

# EDA

```python
bank_data['pep'].value_counts()
```

## Preprocess Data

```python
# one hot encoding sex
bank_data = pd.get_dummies(bank_data, columns=['sex'])
# one hot encoding region
bank_data = pd.get_dummies(bank_data, columns=['region'])
# one hot encoding children
bank_data = pd.get_dummies(bank_data, columns=['children'])
bank_data.head()
```

```python
from scipy.stats import zscore

bank_data['income_zscore'] = zscore(bank_data['income'])

def categorize_income(zscore):
  if zscore < -1:
    return 0
  elif zscore >= -1 and zscore <= 1:
    return 1
  else:
    return 2

bank_data['income_category'] = bank_data['income_zscore'].apply(categorize_income)

bank_data[['income', 'income_zscore', 'income_category']].head()

# one hot encoding income_category and drop income,income_zscore
bank_data = pd.get_dummies(bank_data, columns=['income_category'])
bank_data.drop(['income', 'income_zscore'], axis=1, inplace=True)
bank_data.head()
```

```python
def categorize_age(age):
  if age < 18:
    return 0
  elif age >= 18 and age < 60:
    return 1
  else:
    return 2

bank_data['age_category'] = bank_data['age'].apply(categorize_age)

bank_data[['age', 'age_category']].head()

# one hot encoding age_category and drop age
bank_data = pd.get_dummies(bank_data, columns=['age_category'])
bank_data.drop(['age'], axis=1, inplace=True)
bank_data.head()
```

```python
bank_data['pep'] = bank_data['pep'].map({'YES': True, 'NO': False})
bank_data['married'] = bank_data['married'].map({'YES': True, 'NO': False})
bank_data['car'] = bank_data['car'].map({'YES': True, 'NO': False})
bank_data['save_act'] = bank_data['save_act'].map({'YES': True, 'NO': False})
bank_data['current_act'] = bank_data['current_act'].map({'YES': True, 'NO': False})
bank_data['mortgage'] = bank_data['mortgage'].map({'YES': True, 'NO': False})
bank_data.drop(['id'], axis=1, inplace=True)
bank_data.head()
```

# Generate Assosiaction Rules by using apiori

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
frequent_itemsets=apriori(bank_data,min_support=0.5,use_colnames=True)
# sort frequent_itemsets
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
my_rules = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
my_rules.sort_values(by='confidence', ascending=False)
```
