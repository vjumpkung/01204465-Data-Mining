```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

## Loading data

```python
raw_data = '''TransID,F,G,H,I,J
1,1,1,0,1,1
2,0,1,1,0,1
3,0,1,0,1,1
4,0,1,1,0,1
5,0,1,1,1,1
6,0,1,1,1,0
'''
```

```python
# read csv from raw_data
from io import StringIO
df = pd.read_csv(StringIO(raw_data))
df = df.drop('TransID', axis=1)
df = df.applymap(lambda x: True if x == 1 else False)
df.head()
```

## Generate Frequent Itemset with support

```python
frequent_itemsets=apriori(df,min_support=0.5,use_colnames=True)
frequent_itemsets
```

## 3 Examples of strong association rules

```python
my_rules = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
my_rules.iloc[:3]
```
